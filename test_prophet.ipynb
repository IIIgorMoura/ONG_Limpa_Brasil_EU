```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

# Carregar e preparar os dados
df = pd.read_excel("./.streamlit/total_mensal.xlsx", header=None)
df.columns = df.iloc[0]
df = df.drop(index=0)
df_transposto = df.transpose().reset_index()
df_transposto.columns = ['mes_ano', 'y']
df_transposto['ds'] = pd.to_datetime(df_transposto['mes_ano'], format='%b/%y', errors='coerce')
df_transposto['y'] = pd.to_numeric(df_transposto['y'], errors='coerce')
df_transposto = df_transposto.dropna(subset=['ds', 'y']).sort_values(by='ds').reset_index(drop=True)

# Criar lags (valores anteriores como features)
for lag in range(1, 13):  # 12 lags = 1 ano de histórico
    df_transposto[f'lag_{lag}'] = df_transposto['y'].shift(lag)

# Remover as primeiras linhas com NaN (devido aos lags)
df_model = df_transposto.dropna()

# Separar em variáveis de entrada (X) e saída (y)
X = df_model[[f'lag_{i}' for i in range(1, 13)]]
y = df_model['y']

# Treinar modelo (treinamento até dez/2020)
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)

# Modelo de Random Forest (poderia ser também Ridge, GradientBoosting, etc.)
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Avaliar no teste
y_pred = model.predict(X_test)
import numpy as np
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)

# Previsão futura iterativa (jan/21 a dez/25)
import numpy as np

last_known = df_model['y'].values[-12:]  # últimos 12 valores reais
future_preds = []

for _ in range(58):  # 58 meses = jan/21 a dez/25
    X_future = last_known[-12:].reshape(1, -1)
    next_pred = model.predict(X_future)[0]
    future_preds.append(next_pred)
    last_known = np.append(last_known, next_pred)

# Criar datas futuras
future_dates = pd.date_range(start='2021-01-01', periods=58, freq='MS')
future_df = pd.DataFrame({'ds': future_dates, 'yhat': future_preds})
display(future_df)
# Visualizar
plt.figure(figsize=(12,6))
plt.plot(df_transposto['ds'], df_transposto['y'], label='Histórico')
plt.plot(future_df['ds'], future_df['yhat'], label='Previsão', linestyle='--')
plt.title('Previsão de Lixo Coletado até Dez/2025')
plt.legend()
plt.grid()
plt.show()






# PLOT REAL


# Preparar os dados para o gráfico de comparação
df_transposto['y_pred'] = np.nan  # Inicializa a coluna de previsão com NaN
# df_transposto.loc[df_transposto['ds'] >= future_df['ds'][0], 'y_pred'] = 0  # Zera os valores após a última data do histórico

# Plotando o gráfico com os dados reais até o ponto de previsão
plt.figure(figsize=(12,6))
plt.plot(df_transposto['ds'], df_transposto['y'], label='Histórico', color='blue')
plt.plot(df_transposto['ds'], df_transposto['y_pred'], label='Zeros Após Previsão', linestyle='--', color='red')
plt.plot(future_df['ds'], future_df['yhat'], label='Previsão', linestyle='--', color='green')

plt.title('Histórico e Previsão de Lixo Coletado')
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```
