```python
import pandas as pd

# ALTERAR ANO DE TRATAMENTO, ex: 2013 = 13, 2014 = 14

ano = 21

df = pd.read_csv(f'./DBs/amlurbcoletatiposresiduos20{ano}.csv', encoding='ISO-8859-1', sep=';')
df.to_excel(f'./DBs_tratados/20{ano}_coleta_tipos_residuos.xlsx', index=False)

# transformar em excel
df = pd.read_excel(f'./DBs_tratados/20{ano}_coleta_tipos_residuos.xlsx')

df
```

```python
# converter numPadrão BR > numPadrão US

# PARA OS ANOS 2013, 2021
    # Converter blocos específicos de cada ano

df.info()

colunas_converter = [f'{mes}/{ano}' for mes in ['jan', 'fev', 'mar', 'abr', 'mai', 'jun', 'jul', 'ago', 'set', 'out', 'nov', 'dez']]


for mes_coluna in colunas_converter:
    df = df.fillna(0)
    df[mes_coluna] = df[mes_coluna].astype(str).str.strip().str.replace(',', '.').str.replace('-', '0').astype(float)

df.columns = df.columns.str.strip()
df.rename(columns={"TIPO DE RESÖDUO": "Tipo_Residuo", 'TIPO DE RESÍDUO': 'Tipo_Residuo', 'TIPO DE RESï¿½DUO': 'Tipo_Residuo', 'Tipo de Resíduo - Toneladas': 'Tipo_Residuo', 'Tipo de resï¿½duo - Toneladas': 'Tipo_Residuo', 'Tipo de resíduo - Toneladas': 'Tipo_Residuo', 'Tipo de residuo - Toneladas': 'Tipo_Residuo', f"TOTAL 20{ano}": f"total_20{ano}", f'20{ano}': f'total_20{ano}'}, inplace=True)

df.columns = df.columns.str.lower().str.strip()

df['tipo_residuo'] = df['tipo_residuo'].str.replace('TOTAL', 'total geral')
df['tipo_residuo'] = df['tipo_residuo'].str.lower().str.title()

df = df[df['tipo_residuo'] != '(Vazio)']

# tratamento de anos específicos
if ano == 20 or ano == 21:
    df[f'total_20{ano}'] = df[colunas_converter].sum(axis=1)
else:
    df[f'total_20{ano}'] = df[f'total_20{ano}'].str.strip().str.replace('.', '').str.replace(',', '.').str.replace('-', '0').astype(float)
    df[f'total_20{ano}'] = df[colunas_converter].sum(axis=1)

if ano == 17:
    df = df.drop(columns='unnamed: 14')
elif ano == 24:
    df = df.drop(columns='unnamed: 13')

display(df)
df.info()
```
